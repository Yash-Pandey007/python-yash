```python
class bank:
  def __init__(self,acc_no,acc_holder_name,acc_balance):
    self.acc_no = acc_no
    self.acc_holder_name = acc_holder_name
    self.acc_balance = acc_balance

  def deposit(self,x):
    self.acc_balance += x
  def withdrawal(self,c):
    if self.acc_balance > c:
      self.acc_balance-=c
    else : print("not sufficiant balance")
  def get_balance(self):
    print(self.acc_balance)
```

```python
yash = bank(1234,"yash_pandey",10000)
```

```python
yash.deposit(10000)
```

```python
yash.withdrawal(10000)
```

```python
yash.get_balance()
```

```python
class family:
  def __init__(self,surname,family_income,family_members):
    self.surname = surname
    self.family_income = family_income
    self.family_members = family_members
class member(family):
  def __init__(self,name,age,surname,family_income,family_members):
    super().__init__(surname,family_income,family_members)
    self.name = name
    self.age = age
```

```python
class Stationary:
    def __init__(self,pencil,notebook,pen,eraser,sharpner):
        self.pencil = pencil
        self.notebook=notebook
        self.pen=pen
        self.eraser=eraser
        self.sharpner=sharpner

    def totalbill(self):
        print(self.pencil+self.notebook+self.pen+self.eraser+self.sharpner)

    def totalproduct(self):
        print(self.pencil+self.notebook+self.pen+self.eraser+self.sharpner)
```

```python
shop = Stationary(10,25,15,5,5)
```

```python
shop.totalbill()
```

```python
shop.totalproduct()
```

```python
class Account:
  def __init__(self,balance):
    self.balance = balance

  def interest():
    pass
class saving_account(Account):
  def __init__(self,balance):
    super().__init__(balance)
  def interest(self):
    return self.balance*0.05

class current_account(Account):
  def __init__(self,balance):
    super().__init__(balance)
  def interest(self):
    return self.balance*0.02
```

```python
yash_s = saving_account(10000)
```
