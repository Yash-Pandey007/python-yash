```python
def fact(n):
  if n==0 or n==1:
    return 1
  else:
    return n*fact(n-1)
```

```python
fact(5)
```

```python
def fib(n):
  if n==0:
    return 0
  elif n==1:
    return 1
  else:
    return fib(n-1)+fib(n-2)
```

```python
fib(37)
```
