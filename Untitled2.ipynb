```python
from bs4 import BeautifulSoup as bs
```

```python
import requests as rs
```

```python
url = "https://www.geeksforgeeks.org/courses?source=google&medium=cpc&device=c&keyword=geeksforgeeks&matchtype=e&campaignid=20039445781&adgroup=147845288105&gad_source=1&gclid=Cj0KCQiA6Ou5BhCrARIsAPoTxrD9pkrwmVydxiKMPxO1BCxx_tDN-0jqzpO2ER3otb9V51ODzJKJQIAaArAlEALw_wcB"
```

```python
response=rs.request('get',url).content
```

```python
soup=bs(response,'html.parser')
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
soup_div=soup.find(class_="footer-container_right")
```

```python
print(soup_div.prettify())
```

```python
soup_ul=soup_div.find_all("ul",class_="footer-container_links-list")
```

```python
print(soup_ul)
```

```python
li_tag=[]
for i in soup_ul:
    li_tag.append(i.find_all("li"))
```

```python
li_tag
```

```python
course_name=[]
for i in li_tag:
    for j in i:
        course_name.append(j.text)
```

```python
len(course_name)
course_name = course_name[6:]
```

```python
print(len(course_name))
print(course_name)
```

```python
links=[]
for i in li_tag:
    for j in i:
        for k in j.find_all("a"):
            links.append(k.get("href"))
```

```python
print(links)
```

```python
course={"course_name":course_name,"links":links}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(course)
```

```python
df.to_csv("course.csv")
```
